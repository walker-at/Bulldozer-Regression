## 1. Problem Definition
Given the usage and equipment configurations of a bulldozer, how well can we predict its future sale price?

## Data
The 3 datasets provided by Kaggle are described as:
* Train.csv is the training set, which contains data through the end of 2011.
* Valid.csv is the validation set, which contains data from January 1, 2012 - April 30, 2012 You make predictions on this set throughout the majority of the competition. Your score on this set is used to create the public leaderboard.
* Test.csv is the test set, which won't be released until the last week of the competition. It contains data from May 1, 2012 - November 2012. Your score on the test set determines your final rank for the competition


## Evaluation Metric
The competition evaluates on the root mean squared log error (RMSLE) between the actual and predicted price of a sold bulldozer. We will also be taking a look at the mean absolute error (MAE) and the r^2 score.

### 4.5 Fill missing values
We will be filing numerical values with the median to deal better with outliers. On top of this, we will be adding a column of whether the data was initially missing or not to retain this inofrmation.

```python
df.isna().sum()
```

```python
# fill numeric rows with the median
for label, content in df.items():
  if pd.api.types.is_numeric_dtype(content):
    if pd.isnull(content).sum():
      # add a binary column to indicate if the data was initially missing
      df[label+'_is_missing'] = pd.isnull(content)
      # fill missing numeric values with median
      df[label] = content.fillna(content.median())

# turn categorical variables into numbers and fill missing
for label, content in df.items():
  if not pd.api.types.is_numeric_dtype(content):
    # add a binary column to indicate if the data was initially missing
    df[label+'_is_missing'] = pd.isnull(content)
    # categoreis to  numbers and add 1
    df[label] = pd.Categorical(content).codes + 1
```

### 4.6 Creating a function for preprocessing/transforming data
This function will
1. Turn the 'saledate' feature into more meaningfull datetime features.
2. Drop the original 'saledate' column.
3. Fill missing values in numeric features
4. Turn cateogorical features into numerical values and fill the missing values.
5. Return the trnasformed df

```python
def preprocess_data(df):
  '''
  This function will:
  1. Turn the 'saledate' feature into more meaningfull datetime features.
  2. Drop the original 'saledate' column.
  3. Fill missing values in numeric features
  4. Turn cateogorical features into numerical values and fill the missing values.
  5. Return the trnasformed df
  '''

  df["saleYear"] = df.saledate.dt.year
  df["saleMonth"] = df.saledate.dt.month
  df["saleDay"] = df.saledate.dt.day
  df["saleDayOfWeek"] = df.saledate.dt.dayofweek
  df["saleDayOfYear"] = df.saledate.dt.dayofyear

  df.drop("saledate", axis=1, inplace=True)

  # put 6.4 here?

  # Fill the numeric rows with median
  for label, content in df.items():
      if pd.api.types.is_numeric_dtype(content):
          if pd.isnull(content).sum():
              # Add a binary column which tells us if the data was missing or not
              df[label+"_is_missing"] = pd.isnull(content)
              # Fill missing numeric values with median
              df[label] = content.fillna(content.median())

      # turn cateogircal variables into numbers and fill missing
      if not pd.api.types.is_numeric_dtype(content):
          df[label+"_is_missing"] = pd.isnull(content)
          # We add +1 to the category code because pandas encodes missing categories as -1
          df[label] = pd.Categorical(content).codes+1

  return df
```

## 5. Model Experimentation with RandomForestRegressor


### 5.1 Split data into train/validation sets
Since this is a time-series problem, kaggle splits the data based on dates:
* Training = all samples up until 2011
* Valid = all samples form January 1, 2012 - April 30, 2012
* Test = all samples from May 1, 2012 - November 2012

We only have to split the train and validation set as they came in a single csv file. The test data is available for download spearately.

```python
# split data
df_val = df[df.saleYear == 2012]
df_train = df[df.saleYear != 2012]

len(df_val), len(df_train)
```

```python
# split into x and y
X_train, y_train = df_train.drop('SalePrice', axis=1), df_train['SalePrice']
X_valid, y_valid = df_val.drop('SalePrice', axis=1), df_val['SalePrice']

X_train.shape, y_train.shape, X_valid.shape, y_valid.shape
```

### 5.2 Building a function for multiple evaluation metrics
* MAE - pays attention to brute differences from mean of target values
* RMSLE -  pays more attention to ratios than brute differences. We care more if we are off by 20% than $20
* R^2 -coefficient of determination. Compares model's predictions to the mean of the targets. A score of 0.9 would indicate that 90% of the variance of dependent variable being studied is explainable by the variance of the independent variable.

```python
# competition uses rmsle
def rmsle(y_test, y_preds):
  '''
  calculates root mean squared log error between predicitons and true labels
  '''
  return np.sqrt(mean_squared_log_error(y_test, y_preds))

# create func to eval model on a few different levels
def get_scores(model):
  train_preds = model.predict(X_train)
  val_preds = model.predict(X_valid)
  scores = {'Training MAE': mean_absolute_error(y_train, train_preds),
            'Valid MAE': mean_absolute_error(y_valid, val_preds),
            'Training RMSLE': rmsle(y_train, train_preds),
            'Valid RMSLE': rmsle(y_valid, val_preds),
            'Training R^2': r2_score(y_train, train_preds),
            'Valid R^2': r2_score(y_valid, val_preds)}
  return scores
```

### 5.3 Testing our model on a subset to tune hyperparameters
In an effort to save my computer from processing all 401,125 samples in the dataset, I will be limiting the max_samples to 10,000. This means every n_estimator will only see 40x less samples.

```python
# change max samples value
model = RandomForestRegressor(n_jobs=-1,
                              random_state=33,
                              max_samples=10000)
```

```python
model.fit(X_train, y_train)
```

```python
get_scores(model)
```

### 5.4 Hyperparameter tuning with RandomizedSearchCV

```python
# different RandomForestRegressor hyperparameters
rf_param_grid = {'n_estimators': np.arange(10, 100, 10),
                 'max_depth': [None, 3, 5, 10],
                 'min_samples_split': np.arange(2, 20, 2),
                 'min_samples_leaf': np.arange(1, 20, 2),
                 'max_features': [0.5, 1, 'sqrt', 'auto'],
                 'max_samples': [10000]}

# instantiate
rs_forest_model = RandomizedSearchCV(RandomForestRegressor(n_jobs=-1,
                                                           random_state=42),
                                     param_distributions=rf_param_grid,
                                     n_iter=2,
                                     cv=5,
                                     verbose=True)

rs_forest_model.fit(X_train, y_train)
```

```python
# show best parameters
rs_forest_model.best_params_
```

```python
# evaluate randomizedsearch model
get_scores(rs_forest_model)
```

### 5.5 Train model with the best hyperparameters
These parameters were found after 100 iterations of RandomizedSearchCV across all the data

## 6. Model Experimentation with XGBRegressor for Comparison
Gradient boosting. This time, instead of simultaneously planting a load of independent trees all at once at random (bootstrapping and aggregating aka. bagging), each successive tree that we plant is weighted in such a way as to compensate for any weakness (residual errors) in the previous tree. This is known as boosting.

```python
# Instantiate model
boost_model = XGBRegressor(random_state=42,
                           learning_rate=0.1)

# Fit the model
boost_model.fit(X_train, y_train)
```

### 6.1 Hyperparamter tuning with RandomizedSearchCV

```python
# Define the hyperparameter grid
booster_param_grid = {"n_estimators": st.randint(3, 40),
                      "max_depth": st.randint(3, 40),
                      "learning_rate": st.uniform(0.05, 0.4),
                      "colsample_bytree": st.beta(10, 1),
                      "subsample": st.beta(10, 1),
                      "gamma": st.uniform(0, 10),
                      'reg_alpha': st.expon(0, 50),
                      "min_child_weight": st.expon(0, 50)}

# Instantiate
rs_booster_model = RandomizedSearchCV(XGBRegressor(n_jobs=-1,
                                                   random_state=42),
                                      param_distributions=booster_param_grid,
                                      n_iter=2,
                                      cv=5,
                                      verbose=True)

# Fit
rs_booster_model.fit(X_train, y_train)
```

### 7.2 Feature importance

```python
feature_importances = pd.DataFrame()
feature_importances['Feature'] = X_train.columns
feature_importances['Importance'] = best_forest_model.feature_importances_
feature_importances.sort_values('Importance', ascending=False)[:20]
```

```python
# func for plotting feature importance
def plot_features(columns, importances, n=20):
    df = (pd.DataFrame({'features': columns,
                      'feature importance': importances})
          .sort_values('feature importance', ascending=False)
          .reset_index(drop=True))

    # plot df
    sns.barplot(x="feature importance",
                y="features",
                data=df[:n],
                orient="h")
```

```python
plot_features(X_train.columns, best_forest_model.feature_importances_)
```
